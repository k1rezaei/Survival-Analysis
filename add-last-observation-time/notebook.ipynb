```python
# yyyymmdd --> number
def int_time_to_number(date: int):
    date = str(date)
    return int(date[-2:]) + 30 * int(date[-4:-2]) + 365 * int(date[-8: -4])

# 'yyyy/mm/dd' --> 'yyyymmdd'
def fix_str_time(date: str):
    dates = date.split('/')
    if len(dates) != 3:
        return -1
    if len(dates[0]) == 0:
        return -1
    
    if len(dates[1]) == 0:
        dates[1] = '01'
    if len(dates[2]) == 0:
        dates[2] = '01'
    
    return dates[0] + dates[1] + dates[2]

# 'yyyy/mm/dd' --> number
def str_time_to_number(date: str):
    return int_time_to_number(int(date))
```

```python
date_columns = ['27', '27.1', '28', '28.1', '31', '31.1', '32', '32.1', '14.1', '14.2',
                '17.1', '17.2', '18.1', '18.2', '21.3', '21.4', '22.2', '22.3', '23.2']
```

```python
for index, row in model.iterrows():
    if row['dead'] == 1:
        continue 
    
    #Not dead
    v1, v2 = row['A'], row['diagnosisdate']
    print(v1, v2)
    
    d_time = row['diagnosisdate']
    f_df = df[df['1'] == row['A']]

    fit_row = None
    for index2, row2 in f_df.iterrows():
        if f_df.shape[0] == 1:
            fit_row = row2
            break
        
        if not isinstance(row2['17'], str):
            continue
        
        if str_time_to_number(fix_str_time(row2['17'])) == int_time_to_number(d_time):
            fit_row = row2

    if fit_row is None:
        continue
    
    print(fit_row['1'], fit_row['17'], d_time, "\n----------")
    max_date = ''
    max_number_date = -1
    
    for c in date_columns:
        p_date = fit_row[c]
        if not isinstance(p_date, str) and math.isnan(fit_row[c]):
            continue
        
        
        p_date = fix_str_time(p_date)
        if p_date == -1:
            continue
        
        if str_time_to_number(p_date) > max_number_date:
            max_number_date = str_time_to_number(p_date)
            max_date = p_date
    
    if len(max_date) > 0:
        model.loc[index, 'EZ'] = max_date[0:4] + '/' + max_date[4:6] + '/' + max_date[6:8]
        print(max_date, "\n*****")
    
```

```python
model.to_csv('new_model.csv')
```
